## 📁 Chargement d'une Image

Mets une image dans `../data/sample_images/` et charge-la ici!

```python
from PIL import Image
import os

# Path to sample images
sample_dir = "../data/sample_images/"

# List available images
if os.path.exists(sample_dir):
    images = [f for f in os.listdir(sample_dir) 
              if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp', '.webp'))]
    print(f"📁 Images disponibles ({len(images)}):")
    for img in images:
        print(f"   • {img}")
else:
    print("📁 Le dossier sample_images n'existe pas encore.")
    print("   Ajoute des images pour commencer!")
```

## 📊 Analyse d'Image

Analysons les statistiques visuelles d'une image.

```python
# Choose an image to analyze
# Replace 'your_image.jpg' with your actual image name

test_image = None
if images:
    test_image = os.path.join(sample_dir, images[0])
    
    print(f"🖼️ Image sélectionnée: {images[0]}\n")
    
    # Display the image
    img = Image.open(test_image)
    display(img)
    
    # Get stats
    print("\n📐 Statistiques:")
    stats = get_image_stats(test_image)
    for key, value in stats.items():
        if key != 'path':
            print(f"   • {key}: {value}")
else:
    print("❌ Aucune image à analyser.")
```

## 🎨 Génération ASCII

Transformons l'image en art ASCII!

```python
if test_image:
    # Generate ASCII art
    ascii_art = generate_ascii(test_image, width=80)
    
    # Display
    print("🎨 Art ASCII:\n")
    print(ascii_art)
    
    # Stats
    ascii_stats = get_ascii_stats(ascii_art)
    print(f"\n📏 Dimensions ASCII: {ascii_stats['width']}x{ascii_stats['height']}")
else:
    print("❌ Aucune image disponible.")
```
